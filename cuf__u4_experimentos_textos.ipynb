# Exeperimentos con Textos

```
#@title Importación de librerías

# Primero algunas librerías típicas casi de cualquier código que haga análisis de datos
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# Luego librerías específicas para procesamiento de páginas web y textos html
import bs4 as BeautifulSoup
from urllib.request import urlopen
import requests
import re
```

```
# Dado un artículo se abre la pagina web con requests
url = "https://www.larepublica.pe/economia/"
response = requests.get(url)
html = response.text
```

```
html
```

```
# Luego se traduce el contenido del HTML usando BeautifulSoup
soup = BeautifulSoup.BeautifulSoup(html, 'html.parser')
```

```
soup
```

```
# Luego se encuetra los párrafos tipo "div" que tienen una "clase" que contenga la expresión "article-body"
# La segunda linea hace que se extraiga sólo el texto
article_elements = soup.findAll('h2', {'class': 'CardSectionSubtitle'})
```

```
for articulo in article_elements:
  print(articulo.text)
```
