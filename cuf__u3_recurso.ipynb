## Ejemplo 2: Hacer exploración de la tabla cargada

```python
# Por ejemplo se puede usar los datasets de la libería sklearn
import pandas as pd # librería que permite manipular tablas
from sklearn.datasets import load_iris # libería que permite extraer datasets de un repositorio
```

```python
data_trabajo = pd.DataFrame(load_iris().data)
```

```python
# ahora si se puede imprimir la tabla
data_trabajo
```

```python
# Para mayor claridad se le puede poner encabezados a todas las columnas
data_trabajo.columns =  load_iris().feature_names
data_trabajo.columns
```

```python
data_trabajo
```

## Ejemplo 3: Limpiar datos

```python
# En este caso tenemos la data limpia, pero podría ser que existan celdas no disponibles
# La libería de Pandas tiene una solución para eliminar todas las celdas sin valor
# dropna quiere decir que descarte todas las celdas que no tengan valor (NA)
# axis = 0 quiere decir que descarte en el sentido de las filas
data_limpia = data_trabajo.dropna(axis=0)
data_limpia
```
